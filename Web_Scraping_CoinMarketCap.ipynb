```python
from bs4 import BeautifulSoup as bs
import requests as rq
import re
```

```python
html_completo = rq.get("https://coinmarketcap.com/currencies/anji/")
html_formatado = bs(html_completo.content)
print(html_formatado.prettify())
```

```python
primeiro_h2 = html_formatado.find("h2")
print(primeiro_h2)
```

```python
todos_h2 = html_formatado.find_all("h2")
for ocorrencia_h2 in todos_h2:
  print(ocorrencia_h2.text)
```

```python
tabela = html_formatado.find_all(class_="fmPyWa")
print(tabela)
# pre_requisito = html_formatado.find(id="pre-requisite-section")
# print(pre_requisito)
# --- Basta selecionar / destacar uma linha, um bloco ou algo assim, e então "Ctrl" + "/" e é mágico :)
```

```python
tabela_tr = html_formatado.find_all("tr")
for tabela_tr_lista in tabela_tr:
  print(tabela_tr_lista.text)
```

```python
tabela_td = html_formatado.find_all("td")
for tabela_td_lista in tabela_td:
  print(tabela_td_lista.text)
```

```python
tabela_th = html_formatado.find_all("th")
for tabela_th_lista in tabela_th:
  print(tabela_th_lista.text)
```

```python
texto_td = html_formatado.find_all(class_="sc-1eb5slv-0 hKkaxT")
for texto_td_lista in texto_td:
  print(texto_td_lista)
```

```python
exchange = html_formatado.find_all(class_="sc-1eb5slv-0 hKkaxT")
print(exchange.text)
```

```python
exchange = html_formatado.find("p", class_="sc-1eb5slv-0 hKkaxT")
print(exchange)
```
